```python
f1 = open("Heatmaps sectors different sizes\Setores saída\simulate_riot_sectors_exit_total_difsizes.txt", "r")
f2 = open("Heatmaps sectors different sizes\Setores migração aleatória (0.01)\simulate_riot_sectors_migration_001_total_difsizes.txt", "r")
f3 = open("Heatmaps sectors different sizes\Setores migração aleatória (0.5)\simulate_riot_sectors_migration_05_total_difsizes.txt", "r")
f4 = open("Heatmaps sectors different sizes\Setores migração aleatória e saída (0.01)\simulate_riot_sectors_migration_exit_001_total_difsizes.txt", "r")
f5 = open("Heatmaps sectors different sizes\Setores migração aleatória e saída (0.5)\simulate_riot_sectors_migration_exit_05_total_difsizes.txt", "r")
f6 = open("Heatmaps sectors different sizes\Setores migração aleatória unidirecional e saída (0.01)\simulate_riot_sectors_migration_unidirectional_exit_001_total_difsizes.txt", "r")
f7 = open("Heatmaps sectors different sizes\Setores migração aleatória unidirecional e saída (0.5)\simulate_riot_sectors_migration_unidirectional_exit_05_total_difsizes.txt", "r")
f8 = open("Heatmaps sectors different sizes\Setores migração gregária\simulate_riot_sectors_migration_gregarious_total_difsizes.txt", "r")
f9 = open("Heatmaps sectors different sizes\Setores migração gregária e saída\simulate_riot_sectors_migration_gregarious_exit_total_difsizes.txt", "r")

h_saida = np.zeros((100,100))
h_mig_001 = np.zeros((100,100))
h_mig_05 = np.zeros((100,100))
h_mig_saida_001 = np.zeros((100,100))
h_mig_saida_05 = np.zeros((100,100))
h_mig_uni_saida_001 = np.zeros((100,100))
h_mig_uni_saida_05 = np.zeros((100,100))
h_greg = np.zeros((100,100))
h_greg_saida = np.zeros((100,100))

for _ in range(10000):
    line = f1.readline()
    line = line.split()
    h_saida[int(line[0])][int(line[1])] = float(line[2])
    line = f2.readline()
    line = line.split()
    h_mig_001[int(line[0])][int(line[1])] = float(line[2])
    line = f3.readline()
    line = line.split()
    h_mig_05[int(line[0])][int(line[1])] = float(line[2])
    line = f4.readline()
    line = line.split()
    h_mig_saida_001[int(line[0])][int(line[1])] = float(line[2])
    line = f5.readline()
    line = line.split()
    h_mig_saida_05[int(line[0])][int(line[1])] = float(line[2])
    line = f6.readline()
    line = line.split()
    h_mig_uni_saida_001[int(line[0])][int(line[1])] = float(line[2])
    line = f7.readline()
    line = line.split()
    h_mig_uni_saida_05[int(line[0])][int(line[1])] = float(line[2])
    line = f8.readline()
    line = line.split()
    h_greg[int(line[0])][int(line[1])] = float(line[2])
    line = f9.readline()
    line = line.split()
    h_greg_saida[int(line[0])][int(line[1])] = float(line[2])
    
f1.close()
f2.close()
f3.close()
f4.close()
f5.close()
f6.close()
f7.close()
f8.close()
f9.close()
```
