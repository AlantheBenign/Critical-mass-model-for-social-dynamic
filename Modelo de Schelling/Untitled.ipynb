```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def utility(pho):
    m = 0.5
    if(pho <= 0.5):
        res = 2*pho
    else:
        res = 2*(m-1)*pho + (2 - m)
    return res
```

```python
densities = np.linspace(0.1, 0.9, 9,endpoint=True)

matriz = np.zeros((len(densities), len(densities), 2))
plt.figure()
for i in range(len(densities)):
    for j in range(len(densities)):
        matriz[i][j] = np.array([utility(densities[i]),utility(densities[j])])
        p_1 = min(densities[i], densities[j])
        p_2 = max(densities[i], densities[j])
        if utility(p_1) < utility(p_2):
            color = 'red'
        else:
            color = 'blue'
        plt.scatter(densities[i], densities[j], color=color)

plt.show()
```
