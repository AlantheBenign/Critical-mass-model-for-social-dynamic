```python
import functions as f
import numpy as np
```

```python
file = open('collectedData.txt', 'r')
file.readline()

numClustersAverage = dict()
numClustersDeviation = dict()
clustersSizesAverage = dict()
clustersSizesDeviation = dict()

numClustersAverage_LOG = dict()
numClustersDeviation_LOG = dict()
clustersSizesAverage_LOG = dict()
clustersSizesDeviation_LOG = dict()
listDicts = [numClustersAverage, numClustersDeviation, clustersSizesAverage, clustersSizesDeviation, numClustersAverage_LOG, numClustersDeviation_LOG, clustersSizesAverage_LOG, clustersSizesDeviation_LOG]
    
num = -1
for line in file:
    if 'Average' in line or 'Deviation' in line:
        num += 1
        continue

    elif line == '' or line == '\n':
        continue

    else:
        line = line.split()
        a = float(line[0][1:4])
        b = float(line[1][0:3])
        key = (a,b)
        value = float(line[2]) + 1
        value_log = np.log(float(line[2]) + 1)
        listDicts[num][key] = value
        listDicts[num + 4][key] = value_log

file.close()
```

```python
num = 3

f.generate_heatmap(listDicts[num], 'Number of Clusters Average')
f.generate_heatmap(listDicts[num + 4], 'Log of Number of Clusters Average')
```
